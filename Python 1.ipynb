```python
#In the United States, it's customary to leave a tip for your server after dining in a restaurant, typically an amount equal to 15% or more of your meal's cost. We have most of a tip calculator ready for you, but we need to implement two functions:
"""dollars_to_float: Accepts a string in the format "$##.##," removes the leading "$," and returns the amount as a float (e.g., given "$50.00" as input, it should return 50.0).
percent_to_float: Accepts a string in the format "##%," removes the trailing "%," and returns the percentage as a float (e.g., given "15%" as input, it should return 0.15).
Please assume that the user will input values in the expected formats."""


def main():
    dollars = dollars_to_float(input("How much was the meal?"))
    percent = percent_to_float(input("What percent would you like to tip?"))
    tip = dollars * percent
    print(f"Leave:${tip: .2f}")

def dollars_to_float(d):
    d = float(d.replace("$", " "))
    return d

def percent_to_float(p):
    p = float(p.replace("%", " "))
    p = p/100
    return p

main()
```

```python
#create a program that prompts the user for the answer to the Ultimate Question of Life, the Universe, and Everything, responding with "Yes" if the user inputs 42, "forty-two," or "forty two" (case-insensitive), and "No" otherwise.

x = input("What's the answer to the great question of Life, the Universe and Everything?")
x = x.casefold().strip()
match x:
    case str("42") | "forty two" | "forty-two":
        print("Yes")
    case _:
        print("No")
```

```python
#create a program that prompts the user for a greeting. If the greeting starts with "hello" (case-insensitive), output $0. If it starts with an "h" (but not "hello"), output $20. Otherwise, output $100, while ignoring leading whitespace in the greeting.

greet = input("Greeting")
if greet.casefold().strip().startswith("hello"):
    print("$0")
elif greet.casefold().strip().startswith("h"):
    print("$20")
else:
    print("$100")
```
